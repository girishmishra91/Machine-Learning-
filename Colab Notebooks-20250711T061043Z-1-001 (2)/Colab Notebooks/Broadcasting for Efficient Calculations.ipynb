```python
import numpy as np
```

### 1.Arithmetic

```python
a=np.array([1,2,3])
b=np.array([[1,2,3],[4,5,6]])
print(a)
print(b)
print("_"*20)
# Addition
print(a+[7])
print(a+b)
print("_"*20)

# subtract
print(a-[7])
print(a-b)
print("_"*20)

#Multiplication
print(a*[7])
print(a*b)
print("_"*20)

# Division
print(a/[2])
print(a/b)
print("_"*20)

# Exponential
print(a**[7])
print(a**b)
print("_"*20)

#Modulo
print(a%[7])
print(a%b)
print("_"*20)
```

### 2.Comparison

```python
a=np.array([1,2,3])
b=np.array([[1,2,3],[4,5,6]])
# Greater
print(a>b)
print("_"*20)
#Smaller
print(a<b)
print("_"*20)
# Greater than equal to
print(a>=b)
print("_"*20)
#Smaller than equal to
print(a<=b)
print("_"*20)
# equal to
print(a==b)
print("_"*20)
# not equal to
print(a!=b)
print("_"*20)
```

### 3.Logical

```python
a=np.array([1,2,3])
b=np.array([[1,2,3],[4,5,6]])
print(a)
print(b)
print("-"*20)
print(a&b)
print("-"*20)
print(a|b)
print("-"*20)
print(a^b)
print("-"*20)
print(~a)
print("-"*20)
```

### 4.Bitwise Operator

```python
a=np.array([1,2,3])
b=np.array([[1,2,3],[4,5,6]])
print(a)
print(b)
print("-"*20)
print(np.bitwise_and(a,b))
print("-"*20)
print(np.bitwise_or(a,b))
print("-"*20)
print(np.bitwise_xor(a,b))
print("-"*20)
print(np.bitwise_not(a))
print("-"*20)
print(np.right_shift(a,b))
print("-"*20)
print(np.left_shift(a,b))
```
