```python
import numpy as np
```

## 1.Saving the txt file

```python
arr=np.array([[1,2],[3,4]])
np.savetxt('data.txt',arr)
```

## 2.npy

```python
arr=np.array([[1,2],[3,4]])
np.save('data.npy',arr)
```

## 3.Csv

```python
arr=np.array([[1,2],[3,4]])
np.savetxt('data.csv',arr,delimiter=',',header='a,b,c')
```

## 4.json

```python
arr=np.array([[1,2,3],[4,5,6]])
np.savetxt('data.json',arr,delimiter=',',header='{"data" :[')
np.savetxt('data.json',arr,delimiter=',',footer=']}')
```

## 5.yaml

```python
arr=np.array([[1,2,3],[4,5,6]])
np.savetxt('data.yaml',arr,delimiter=',',header='-data :')
```

# READING THE DATA NUMPY

### 1.textfile

```python
data=np.loadtxt('data.txt')
print(data)
```

### 2.npy

```python
data=np.load('data.npy')
print(data)
```

### 3.csv

```python
data = np.loadtxt('data.csv', delimiter=',')
print(data)
```

### 4.json

```python
data=np.loadtxt('data.json',delimiter=',')
print(data)
```

### 5.yaml

```python
data=np.loadtxt('data.yaml',delimiter=',')
print(data)
```

#READING DATA WITH NUMPY

### 1.Replacing existing content

```python

data=np.array([1,2,3])
np.savetxt('data-1.txt',data,delimiter=',')
data=np.array([4,5,6])
np.savetxt('data-2.txt',data,delimiter=',')
```

### 2.Appending the data on existing file

```python

data=np.array([1,2,3])
np.savetxt('data.txt',data,delimiter=',')
data=np.load('data.npy')
print(data)
```
