

*   Import all the required  libraries





```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

Load the data

```python
data=pd.read_csv("/content/winequalityN.csv")
data.head()
```

```python
data.columns
```

Note:EDA Implementation is skipped in this pipeline



*   Data Splitting:Split the data into input features and target feature



```python
# x= input feature
x=data.drop(columns='quality',axis=1)
# y=target feature
y=data['quality']
```



*   Data Splitting :Split the data into train and test set



```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=44)
```

##### Feature Scaling :StandardScaler,convert the entire feature set info standard normal form .It simply means that mean=0 & std=1

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x_train = sc.fit_transform(x_train)
x_test = sc.transform(x_test)
```

#### Data Modelling:SVC Classifier

```python
from sklearn.svm import SVC

svm_classifier = SVC()
svm_classifier.fit(x_train, y_train)
predictions = svm_classifier.predict(x_test)
```

```python
svm_classifier.score(x_test,y_test)
```



*   Hyperparameter tuning:GridSearchCV or RandomizedSearchCV




```python
param_grid = {
    'C': [0.1, 1, 10],
    'kernel': ['linear', 'rbf', 'poly'],  # fixed typo here
    'gamma': ['scale', 'auto'],
    'degree': [2, 3, 4]  # used only when kernel='poly'
}

from sklearn.model_selection import GridSearchCV
grid_search = GridSearchCV(svm_classifier, param_grid, cv=5, n_jobs=-1, verbose=2)
grid_search.fit(x_train, y_train)

print("Best parameters found:", grid_search.best_params_)
print("Best cross-validation score:", grid_search.best_score_)
```
